```python
import pandas as pd
import os
```

```python
files = os.listdir("data/")
files = [file for file in files if file.endswith(".xlsx")]
vendas_ds = []
for file in files:    
    data = pd.read_excel(f'data/{file}')
    vendas_ds.append(data)
vendas_ds = pd.concat(vendas_ds).drop_duplicates()
vendas_ds.head()
```

```python
vendas_ds.to_csv('data/base_vendas.csv', index=False)
```
